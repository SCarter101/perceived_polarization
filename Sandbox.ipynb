```python
import PerceivedPolarization
```

```python
# Create Social Network 
socialnetwork = PerceivedPolarization.SocialNetwork(population=20,topology="complete")

# Create Polarization Model
model = PerceivedPolarization.PerceivedPolarization(socialnetwork)
model.global_perceived_opinion()

# Draw graph
model.draw_graph()

# Print final opinions
for node in model.G.nodes:
    print("Node ", node, " personal opinion: ", model.G.nodes[node]['personal_opinion'], " perceived opinion: ", model.G.nodes[node]['perceived_opinion'])
```
